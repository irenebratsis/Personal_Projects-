```python
sum_column = mardf_1["Rent"] - mardf_1["Market_Rent"]
mardf_1["col3"] = sum_column
print(mardf_1)
```

```python

plt.hist(mardf_1.col3)
plt.title("The distribution of Rent minus Market Rent")
plt.xlabel("MarketRentChange")
plt.ylabel("number of occurrence")
plt.show()
```
