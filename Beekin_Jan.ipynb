```python


indexNames = jandf_1[jandf_1['Market_Rent'] == '5.34%'].index
jandf_1.drop(indexNames , inplace=True)
```

```python
jandf_1["Market_Rent"] = pd.to_numeric(jandf_1["Market_Rent"])
```

```python
jandf_1.dtypes
```

```python
sum_column = jandf_1["Rent"] - jandf_1["Market_Rent"]
jandf_1["col3"] = sum_column
print(jandf_1)
```

```python

plt.hist(jandf_1.col3)
plt.title("The distribution of Rent minus Market Rent")
plt.xlabel("Rent minus Market Rent")
plt.ylabel("number of occurrence")
plt.show()
```

```python
jandf_1["col3"].plot()
```

```python
plt.boxplot([jandf_1.col3])
plt.xticks([1,2,3], ['col3'])
plt.show()
```

```python
from scipy import stats
jandf_1['z_score']=stats.zscore(jandf_1['col3'])

jandf_1.loc[jandf_1['z_score'].abs()<=3]
```
